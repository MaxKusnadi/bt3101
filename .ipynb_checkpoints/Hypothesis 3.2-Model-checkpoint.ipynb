```python
storeAssort = data.StoreAssortment.unique()
```

```python
densityList = data.PromoFrequencyRatio.unique()
```

```python
dataset = {}
```

```python
def populateDataset(elem):
    dataset[elem] = {}
    dataset[elem]['subset'] = data.loc[lambda x: x.StoreAssortment == elem]
    dataset[elem]['df'] = pd.DataFrame(columns=['Ratio', 'Sales'])
```

```python
for elem in storeAssort:
    populateDataset(elem)
```

```python
dataset
```

```python
import math
for density in densityList:
    for k, v in dataset.items():
        sls = v['subset'].loc[lambda x: x.PromoFrequencyRatio == density].Sales.sum()
        if(not math.isnan(sls)):
            sls = sls / v['subset'].loc[lambda x: x.PromoFrequencyRatio == density].Sales.count()
        row = {'Ratio': density,
               'Sales': sls}
        v['df'] = v['df'].append(row, ignore_index=True)
```

```python
sales_min_max = {}
```

```python
def normalize_sales(c):
    sales = c.Sales
    store = str(int(c.Store))
    if store in sales_min_max:
        minSales = sales_min_max[store]['minSales']
        maxSales = sales_min_max[store]['maxSales']
    else:
        minSales = data.loc[lambda x: x.Store == int(store)].Sales.min()
        maxSales = data.loc[lambda x: x.Store == int(store)].Sales.max()
        sales_min_max[store] = {}
        sales_min_max[store]['minSales'] = minSales
        sales_min_max[store]['maxSales'] = maxSales
    return (sales - minSales) / (maxSales-minSales)
```

```python
data['NormalizedSales'] = data.apply(lambda x: normalize_sales(x), axis=1)
```
